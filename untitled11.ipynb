```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1=pd.read_csv('D:\Python\\googleplaystore-App-data.csv')
df2=pd.read_csv('D:\Python\\googleplaystore_user_reviews.csv')
```

```python
df1['Installs']=df1.Installs.apply(lambda x: x.replace('+',''))
df1['Installs']=df1.Installs.apply(lambda x: x.replace(',',''))
df1['Installs']=df1.Installs.apply(lambda x: x.replace('Free','0'))
df1['Installs']=df1.Installs.apply(lambda x: x.replace('0.00','0'))
df1['LastUpdated']=pd.to_datetime(df1['LastUpdated'])
```

```python
df1['Installs']=df1['Installs'].astype(int)
df1.dtypes
```

```python
df20= pd.DataFrame(columns=[])
df20['year'] = df1['LastUpdated'].dt.year
df20['month'] = df1['LastUpdated'].dt.month
df20['date'] = df1['LastUpdated'].dt.date
df20['Category'] = df1['Category']
df20['Installs'] = df1['Installs']
df20['Installs'].astype(int)
```

```python
df20['Qtr'] = pd.PeriodIndex(pd.to_datetime(df20.date), freq='Q')
df20.dtypes
```

```python
df37=df20.groupby('Qtr')[['Installs']].sum()
df37.columns
df37.reset_index(inplace=True)
print(df37.dtypes)
df37['Installs'].astype(int)
df37
```

```python
#reuirement11
plt.figure(1,figsize=(15,10))
plt.xticks(rotation='vertical')
sns.set(font_scale=1)
plt.ticklabel_format(style='plain', axis='y')
sns.barplot(x=df37['Qtr'],y=df37['Installs'])
plt.savefig('avg_rating.jpg')
```

```python
#requiremnt18
from csv import writer
def append_list_as_row(file_name, list_of_elem):
    # Open file in append mode
    with open(file_name, 'a+', newline='') as write_obj:
        # Create a writer object from csv module
        csv_writer = writer(write_obj)
        # Add contents of list as last row in the csv file
        csv_writer.writerow(list_of_elem)
```
