```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('D:\Python\\googleplaystore-App-data.csv')
df2=pd.read_csv('D:\Python\\googleplaystore_user_reviews.csv')
```

```python
df1['Installs']=df1.Installs.apply(lambda x: x.replace('+',''))
df1['Installs']=df1.Installs.apply(lambda x: x.replace(',',''))
df1['Installs']=df1.Installs.apply(lambda x: x.replace('Free','0'))
df1['Installs']=df1.Installs.apply(lambda x: x.replace('0.00','0'))
df1['LastUpdated']=pd.to_datetime(df1['LastUpdated'])
```

```python
df1['Installs']=df1['Installs'].astype(int)
```

```python
df1
```

```python
def plotgraphsentiment(App_name,df):
    df50=df[df['App']==App_name]
    df6=df50[df50['Sentiment']=='Positive']
    sum1=df6['Sentiment'].count()
    count1=[]
    count1.append(sum1)
    df7=df50[df50['Sentiment']=='Negative']
    sum1=df7['Sentiment'].count()
    count1.append(sum1)
    df8=df50[df50['Sentiment']=='Neutral']
    sum1=df8['Sentiment'].count()
    count1.append(sum1)
    labels1=['Positive','Negative','Neutral']
    df111=pd.DataFrame(labels1,count1)
    df111.reset_index(inplace=True)
    df111.columns=['Count','Review']
    df111=df111[['Review','Count']]
    plt.figure(1,figsize=(4,4))
    sns.barplot(y=df111['Review'],x=df111['Count'])
    print(df111.head())
```

```python
plotgraphsentiment('104 找工作 - 找工作 找打工 找兼職 履歷健檢 履歷診療室',df2)
```

```python
plotgraphsentiment('1800 Contacts - Lens Store',df2)
```

```python
def maximumdownloads(df12,category):
    df46=df12[df12['Category']==category]
    df20= pd.DataFrame(columns=[])
    df20['year'] = df46['LastUpdated'].dt.year
    print(df20['year'].unique())
    df20['month'] = df46['LastUpdated'].dt.month
    df20['date'] = df46['LastUpdated'].dt.date
    df20['Category'] = df46['Category']
    df20['Installs'] = df46['Installs']
    df20['Installs'].astype(int)
    df20=df20.groupby('month')[['Installs']].sum()
    print(df20[df20['Installs']==df20['Installs'].max()])
    
```
